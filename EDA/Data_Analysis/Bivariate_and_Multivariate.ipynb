```python
import seaborn as sns
import pandas as pd
```

```python
tips= sns.load_dataset('tips')
```

```python
flights=sns.load_dataset('flights')
```

```python
titanic=pd.read_csv('train.csv')
```

```python
iris=sns.load_dataset('iris')
```

# **Scatter Plot (Numerical-Numerical Plot)**


```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
#Multivariate
sns.scatterplot(x='total_bill',y='tip',hue='sex',style='smoker',size='size',data=tips)
```

#**Bar Plot (Numerical-Categorial)**

```python
sns.barplot(x='Pclass',y='Age',data=titanic)
# MULTIVARIATE

sns.barplot(x='Pclass',y='Age',hue='Sex',data=titanic)
```

#**Box Plot (Numerical-Categorial)**

```python
#sns.boxplot(x='Sex',y='Age',data=titanic)
#Multivariate
sns.boxplot(x='Sex',y='Age',hue='Survived',data=titanic)
```

#**DisPlot (Numerical-Categorial)**

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'])
sns.distplot(titanic[titanic['Survived']==1]['Age'])
```

#**HeatMap Categorical-Categorical**


```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
titanic.head()
```

```python
titanic.dropna(subset=['Pclass', 'Survived'], inplace=True)
titanic['Pclass'].fillna(0, inplace=True)
titanic['Survived'].fillna(0, inplace=True)
```

```python
(titanic.groupby('Pclass').mean(numeric_only=True)['Survived']*100).plot(kind='bar')
```

#**ClusterMap Categorical-Categorical**


```python
sns.clustermap(pd.crosstab(titanic['SibSp'],titanic['Survived']))
```

#**PairPlot**

```python
sns.pairplot(iris,hue='species')
```

# **Line plot (Numerical-Numerical)**

```python
#special case of line plot
new=flights.groupby('year').sum(numeric_only=True).reset_index()
```

```python
sns.lineplot(x='year',y='passengers',data=new)
```

```python
sns.heatmap(flights.pivot_table(values='passengers',index='year',columns='month'))
```

```python
sns.clustermap(flights.pivot_table(values='passengers',index='year',columns='month'))
```
