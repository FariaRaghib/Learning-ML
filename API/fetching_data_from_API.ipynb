```python
df=pd.DataFrame()
```

```python
for i in range(1,483):
  response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=2f6e8b607f96604b9eb90da8c6b0fc82&language=en-US&page={}'.format(i))
  dfx=pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','vote_count','vote_average','popularity']]
  df=pd.concat([df,dfx],ignore_index=True)
```

```python
df.to_csv('movies.csv')
```
