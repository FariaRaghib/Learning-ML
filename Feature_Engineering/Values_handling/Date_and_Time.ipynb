```python
date['date']=pd.to_datetime(date['date'])
```

```python
date.info()
```

# Extract year

```python
date['date_year']=date['date'].dt.year
```

```python
date.head()
```

# Extract Month

```python
date['date_month']=date['date'].dt.month
```

```python
date.head()
```

```python
date['date_month_name']=date['date'].dt.month_name()

date.head()
```

```python
date['date_day']=date['date'].dt.day
date.head()
```

```python
date['date_dayofweek']=date['date'].dt.dayofweek
date.head()
```

```python
date['date_name']=date['date'].dt.day_name()
date.head()
```

```python
date['date_is_weekend']=np.where(date['date_name'].isin(['Sunday','Saturday']),1,0)
date.head()
```

```python
date['date_week']=date['date'].dt.isocalendar().week
date.head()
```

```python
date['date_quarter']=date['date'].dt.quarter
date.head()
```

```python
date['date_semester']=np.where(date['date_quarter'].isin([1,2]),1,2)
date.head()
```

```python
import datetime

today=datetime.datetime.today()

today
```

```python
today-date['date']
```

```python
(today-date['date']).dt.days
```

```python

# Months passed

np.round((today -date['date']) / np.timedelta64(1, 'D'),0)
```
